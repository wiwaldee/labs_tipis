```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

# Определите функцию f(x)
def f(x):
    return x**4 - x - 1

# Определите производную функции f(x)
def f_prime(x):
    return 4*x**3 - 1

# Начальные приближения для поиска экстремумов
initial_guesses = [-1.0, 1.0]

extrema = []  # Здесь будут храниться точки экстремумов

# Выполните численную оптимизацию для каждого начального приближения
for guess in initial_guesses:
    root = newton(f, guess, f_prime)
    extrema.append(root)

# Выведите найденные точки экстремумов
for i, root in enumerate(extrema):
    print(f'Экстремум {i + 1}: x = {root:.5f}, f(x) = {f(root):.5f}')

# Постройте график функции
x = np.linspace(-2, 2, 400)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x^4 - x - 1', color='b')
plt.scatter(extrema, [f(root) for root in extrema], color='r', label='Точки экстремумов', zorder=5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.title('График функции и точки экстремумов')
plt.show()
```

```python
import math

# Определение функции f(x)
def f(x):
    return x**4 - x - 1

# Определение функции для чисел Фибоначчи
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        a, b = 0, 1
        for _ in range(2, n + 1):
            a, b = b, a + b
        return b

# Нахождение интервала, в котором находится максимальное значение f(x)
epsilon = 1e-5  # Заданная точность
a = -2.0  # Левая граница интервала
b = 2.0  # Правая граница интервала

# Вычисление числа итераций для чисел Фибоначчи
n = 2
while (b - a) / fibonacci(n) >= epsilon:
    n += 1

# Найденное значение n определяет, сколько раз нужно разделить интервал

# Применение метода золотого сечения с использованием чисел Фибоначчи
for _ in range(n - 1):
    x1 = a + (b - a) * (fibonacci(n - 2) / fibonacci(n))
    x2 = a + (b - a) * (fibonacci(n - 1) / fibonacci(n))

    f_x1 = f(x1)
    f_x2 = f(x2)

    if f_x1 > f_x2:
        b = x2
    else:
        a = x1

# Точка, в которой достигается максимальное значение f(x)
x_max = (a + b) / 2
max_value = f(x_max)

print(f'Наибольшее значение функции f(x) равно {max_value:.5f} при x = {x_max:.5f}')
```

```python
# Определение функции f(x) и её производной f'(x)
def f(x):
    return x**4 - x - 1

def f_prime(x):
    return 4 * x**3 - 1

# Начальное значение x0 и параметры метода
x0 = 1.0  # Начальное значение
learning_rate = 0.1  # Размер шага
max_iterations = 1000  # Максимальное количество итераций
tolerance = 1e-6  # Точность для остановки

# Итеративное обновление x с использованием метода наискорейшего спуска
for iteration in range(max_iterations):
    gradient = f_prime(x0)
    x1 = x0 - learning_rate * gradient

    # Проверка условия остановки
    if abs(x1 - x0) < tolerance:
        break

    x0 = x1

min_value = f(x0)
print(f'Наименьшее значение функции f(x) равно {min_value:.5f} при x = {x0:.5f}')
```
