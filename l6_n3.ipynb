```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Определим последовательности
def a_n(n):
  return 1/(np.power(n, 1+(1/n)))

def b_n(n):
  return (np.power((-1), n)) * (np.power(n, 1/2)) - (np.power((n-1), 1/2))
```

```python
# Задание 1
n_values = np.arange(1, 101)
a_values = [a_n(n) for n in n_values]
b_values = [b_n(n) for n in n_values]

# Построим последовательности
plt.figure(figsize=(10, 5))
plt.plot(n_values, a_values, label='a_n')
plt.plot(n_values, b_values, label='b_n')
plt.xlabel('n')
plt.ylabel('Значение')
plt.legend()
plt.title('Последовательность')
plt.grid(True)
plt.show()


# Оценка пределов
print(f"Примерный предел a_n: {np.mean(a_values[-10:]):.5f}")
print(f"Примерный предел b_n: {np.mean(b_values[-10:]):.5f}")


# Задание 2
epsilon = 1e-4
n_values = np.arange(1, 10001)
a_values = [a_n(n) for n in n_values]
partial_sums = np.cumsum(a_values)
difference = np.diff(partial_sums)
diverging = any(abs(val) < epsilon for val in difference[-100:])

if diverging:
    print("Ряд a_n расходится.")
else:
    print(f"Сумма серий для a_n до n = 10000: {partial_sums[-1]:.5f}")
```
